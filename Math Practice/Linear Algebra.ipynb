Question: Which vectors $V\in \mathbb{R}^2$ are perpendicular to the vector $u=\begin{pmatrix}2\\1\end{pmatrix}$?


A vector $\vec{v}$ is perpendicular to $u$ if $\vec{u}\cdot \vec{v} = 0$

$\begin{bmatrix}\vec{v}_1\\ \vec{v}_2\end{bmatrix} \cdot \begin{bmatrix}2\\ 1\end{bmatrix} = 0$

$2\vec{v}_1 + 1\vec{v}_2 = 0$ -- right? \
$2 + 1 + \vec{v}_1 + \vec{v}_2$ = 0 -- wrong?\
$1\vec{v}_1 + \frac{1}{2}\vec{v}_2 = 0$\
$\begin{bmatrix}1\\\frac{1}{2}\end{bmatrix}=0$

$\vec{v}_1 + \vec{v}_2 = -3$
